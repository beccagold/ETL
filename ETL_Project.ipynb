```python
import pandas as pd
import numpy as np
from sqlalchemy import create_engine
```

```python
ps4_sales = pd.read_csv("videogames_data/PS4_GamesSales.csv")
other_sales = pd.read_csv("videogames_data/Video_Games_Sales_as_at_22_Dec_2016.csv")
xbox_sales = pd.read_csv("videogames_data/XboxOne_GameSales.csv")
```

```python
ps4_df = ps4_sales[['Game', 'Year','Genre', 'North America','Europe']].dropna()
```

```python
xbox_df = xbox_sales[['Game','Year','Genre', 'North America','Europe']].dropna()
len(xbox_df['Game'])
```

```python
other_df = other_sales[['Name','Year_of_Release','Genre','NA_Sales','EU_Sales']].rename(columns={'Name':'Game',"Year_of_Release":'Year','NA_Sales':'North America', 'EU_Sales':'Europe'}).dropna()
other_df
```

```python
#vg_df1 = ps4_df.merge(xbox_df, how = 'inner', on ='Game')
#vg_df1['NA'] = vg_df1['North_America_x']+vg_df1['North_America_y']
#vg_df = vg_df1.dropna().merge(vg_df1, other_df, how = 'inner', on = vg_df1['Game'])
vg = pd.concat([ps4_df,xbox_df])
vg_pd = pd.concat([vg, other_df])
#len(vg_pd['Game'])
vg_pd.groupby('Game').head()
```

```python
csv = 'shootings_data/Mass_Shootings_Dataset.csv'
ms_df = pd.read_csv(csv)
ms_df.head()
```

```python
ms_db = ms_df[['Title', 'Location', 'Date', 'Fatalities', 'Injured', 'Total victims', 'Race']]
ms_db.head()
```

```python
ms_columns = ['Title', 'Location', 'Date', 'Fatalities', 'Injured', 'Total victims', 'Race']
ms_db_transformed = ms_db[ms_columns].copy()

ms_df = ms_db_transformed.rename(columns={"Title": "title",
                                        "Location": "location",
                                        "Date": "date",
                                        "Fatalities": "fatalities",
                                        "Injured": "injured",
                                        "Total victims": "total_victims",
                                        "Race": "race"})

ms_df.set_index('title', inplace=True)
ms_df.head()
```

```python
vg_columns = ['Game', 'Year', 'Genre', 'North America', 'Europe']
vg_pd_transformed = vg_pd[vg_columns].copy()

vg_pd_transformed = vg_pd_transformed.rename(columns={'Game': 'game',
                                                     'Year': 'year',
                                                     'Genre': 'genre',
                                                     'North America': 'north_america',
                                                     'Europe': 'europe'})
vg_pd_transformed.set_index('game', inplace=True)
vg_pd_transformed.head()
```

```python
from sqlalchemy_utils import create_database

engine = create_engine(f'postgresql://postgres:Erau2019!@localhost:5432/vg_pd')
create_database(engine.url)

ms_df.to_sql('ms_df', con=engine, if_exists='append', index=True)
vg_pd_transformed.to_sql('vg_pd', con=engine, if_exists='append', index=True)
```
